# display_tea_iou_result

teaの結果を表示するノートブック

```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
reslut_dic_en_save_path = '/home/nakachi/data/category_tea_iou_en_20220112.pk'
reslut_dic_ja_save_path = '/home/nakachi/data/category_tea_iou_ja_20220112.pk'
```

```python
with open(reslut_dic_en_save_path, 'rb') as f:
    tea_result_dic_en = pickle.load(f)

with open(reslut_dic_ja_save_path, 'rb') as f:
    tea_result_dic_ja = pickle.load(f)
```

```python
en_tea_iou_list = []
en_tea_iou_std_list = []
ja_tea_iou_list = []
ja_tea_iou_std_list = []

for category in tea_result_dic_en.keys():
    print('-----------------------------------------')
    print(category)
    print(f'English TEA-IoU: {tea_result_dic_en[category]["tea_iou"]}')
    print(f'English TEA-IoU std: {tea_result_dic_en[category]["std_tea_iou"]}')
    print(f'Accpete rate: {tea_result_dic_en[category]["num_calu_iou"] / tea_result_dic_en[category]["num_sentence"]}')
    print(f'Japanese TEA-IoU: {tea_result_dic_ja[category]["tea_iou"]}')
    print(f'Japanese TEA-IoU std: {tea_result_dic_ja[category]["std_tea_iou"]}')
    print(f'Accpete rate: {tea_result_dic_ja[category]["num_calu_iou"] / tea_result_dic_ja[category]["num_sentence"]}')
    
    en_tea_iou_list.append(tea_result_dic_en[category]['tea_iou'])
    en_tea_iou_std_list.append(tea_result_dic_en[category]['std_tea_iou'])
    ja_tea_iou_list.append(tea_result_dic_ja[category]['tea_iou'])
    ja_tea_iou_std_list.append(tea_result_dic_ja[category]['std_tea_iou'])

print('-----------------------------------------')
print('English total mean:', sum(en_tea_iou_list)/len(en_tea_iou_list))
print('English total mean std:', sum(en_tea_iou_std_list)/len(en_tea_iou_std_list))
print('Japanese total mean:', sum(ja_tea_iou_list)/len(ja_tea_iou_list))
print('Japanese total mean std:', sum(ja_tea_iou_std_list)/len(ja_tea_iou_std_list))
```

```python
# error barあり：没
labels = []
en_tea_iou_list = []
en_tea_iou_std_list = []
ja_tea_iou_list = []
ja_tea_iou_std_list = []
for category in tea_result_dic_en.keys():
    labels.append(category)
    en_tea_iou_list.append(tea_result_dic_en[category]['tea_iou'])
    en_tea_iou_std_list.append(tea_result_dic_en[category]['std_tea_iou'])
    if category in tea_result_dic_ja:
        ja_tea_iou_list.append(tea_result_dic_ja[category]['tea_iou'])
        ja_tea_iou_std_list.append(tea_result_dic_ja[category]['std_tea_iou'])
    else:
        ja_tea_iou_list.append(0.0)
        ja_tea_iou_std_list.append(0.0)

labels = labels[:20]
left = np.arange(len(en_tea_iou_list[:20]))
width = 0.3

plt.figure(figsize=(30,10))
plt.bar(left, en_tea_iou_list[:20], yerr=en_tea_iou_std_list[:20], color='r', width=width, align='center', label='coco')
plt.bar(left+width, ja_tea_iou_list[:20], yerr=ja_tea_iou_std_list[:20], color='b', width=width, align='center', label='stair')

plt.xticks(left + width/2, labels, fontsize=20)
plt.legend(fontsize=20)
plt.show()
```

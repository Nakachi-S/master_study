# compare_ja_en_eval
指定したカテゴリー評価データセットの、それぞれ生成した画像を比較するノート。

全evalを生成した画像の中から引っ張り出すだけ。

```python
import pickle
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
import os
```

```python
val_filename_path = '/home/nakachi/data/category_split/snowboard_en/val_filenames.pickle'
with open(val_filename_path, 'rb') as f:
    val_filenames = pickle.load(f)
```

```python
JA_PATH = '/home/nakachi/data/stair/AttnGAN/output/stair_attn2-stair_2021_02_04_10_23_01/Model/netG_epoch_120/valid/single'
EN_PATH = '/home/nakachi/data/models/batch_is_12_coco_AttnGAN2/valid/single/'
MAX_DISPLAY = 5
```

```python
%matplotlib inline
for i, file_name in enumerate(val_filenames):
    if i == MAX_DISPLAY:
        break
    
    ja_image_path = os.path.join(JA_PATH, file_name + '_s-1.png')
    en_image_path = os.path.join(EN_PATH, file_name + '_s-1.png')
    
    ja_img = mpimg.imread(ja_image_path)
    en_img = mpimg.imread(en_image_path)
    
    plt.subplot(MAX_DISPLAY, 2, i*2+1)
    plt.imshow(ja_img)
    
    plt.subplot(MAX_DISPLAY, 2, i*2+2)
    plt.imshow(en_img)

    

plt.show()
```

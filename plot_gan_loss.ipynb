# plot_gan_loss

ganのlossをプロットする

```python
import matplotlib.pyplot as plt
import json
import os
```

```python
EN_MODEL_PATH = '/home/nakachi/data/stair/AttnGAN/output/coco_attn2_2022_01_26_17_14_43/Model'
JA_MODEL_PATH = '/home/nakachi/data/stair/AttnGAN/output/stair_attn2_2022_01_26_18_00_41/Model'
```

```python
# 英語
g_loss = []
d_loss = []
for i in range(120):
    epoch_loss_path = os.path.join(EN_MODEL_PATH, f'loss_epoch_{i}.json')
    with open(epoch_loss_path) as f:
        epoch_loss = json.load(f)
        g_loss.append(epoch_loss['LossG'])
        d_loss.append(epoch_loss['Loss_D'])

        
# データの定義(サンプルなのでテキトー)
x = list(range(120))

# グラフの描画
plt.plot(x, g_loss)
plt.show()
```

```python
plt.plot(x, d_loss)
plt.show()
```

```python
import torch
import torchvision.transforms as transforms
from PIL import Image, ImageDraw, ImageFont
import random
import cv2
import skimage.transform
from matplotlib.collections import PatchCollection
from matplotlib.patches import Polygon

def polygon_resize(image, segmentation, width, height):
    # 現在の高さと幅を取得しておく
    c_height, c_width = image.shape[:2]
    img = skimage.transform.resize(image, (width, height))

    # 圧縮する比率(rate)を計算
    r_width = width / c_width
    r_height = width / c_height

    # 比率を使ってBoundingBoxの座標を修正
    new_segmentation = []
    for seg in segmentation:
        new_seg = []
        for i in range(0, len(seg), 2):
            new_seg.append(seg[i] * r_width)
            new_seg.append(seg[i+1] * r_height)
        new_segmentation.append(new_seg)
    return img, new_segmentation
```

```python
INSTA_VAL_ANNOTATIONS_PATH = "/home/nakachi/data/coco2014/annotations/instances_val2014.json"
insta_val_coco = COCO(INSTA_VAL_ANNOTATIONS_PATH)
img_ids = 546622
annotation_ids = insta_val_coco.getAnnIds(imgIds=img_ids, catIds=[17], iscrowd=None)
# print(annotation_ids)
annotations = insta_val_coco.loadAnns(annotation_ids)
print(annotations)

for ann in annotations:
    image_meta = insta_val_coco.loadImgs(ann["image_id"])[0]
    image_path = os.path.join(VAL_IMAGES_DIRECTORY, image_meta["file_name"])
#     print(ann)
    I = io.imread(image_path)

    new_img, new_segmentation = polygon_resize(I, ann["segmentation"], 256, 256)
    plt.imshow(new_img)
    ann['segmentation'] = new_segmentation
    print(ann)
    insta_val_coco.showAnns([ann])
```

```python
class Box:
    def __init__(self, x, y, w, h):
        self.x = x
        self.y = y
        self.w = w
        self.h = h

def resize(image, boxes, width, height):
    # 現在の高さと幅を取得しておく
    c_height, c_width = image.shape[:2]
    img = skimage.transform.resize(image, (width, height))

    # 圧縮する比率(rate)を計算
    r_width = width / c_width
    r_height = width / c_height

    # 比率を使ってBoundingBoxの座標を修正
    new_boxes = []
    for box in boxes:
        x = int(box.x * r_width)
        y = int(box.y * r_height)
        w = int(box.w * r_width)
        h = int(box.h * r_height)
        new_box = Box(x, y, w, h)
        new_boxes.append(new_box)
    return img, new_boxes
```

```python
bbox = Box(annotations[0]['bbox'][0], annotations[0]['bbox'][1], annotations[0]['bbox'][2], annotations[0]['bbox'][3])
```

```python
new_img, new_boxes = resize(img, [bbox], 256, 256)
```

```python
from matplotlib.collections import PatchCollection
from matplotlib.patches import Polygon

# bboxの値を決める
[bbox_x, bbox_y, bbox_w, bbox_h] = new_boxes[0].x, new_boxes[0].y, new_boxes[0].w, new_boxes[0].h

polygons = []
color = []
c = (np.random.random((1, 3))*0.6+0.4).tolist()[0]
poly = [[bbox_x, bbox_y], [bbox_x, bbox_y+bbox_h], [bbox_x+bbox_w, bbox_y+bbox_h], [bbox_x+bbox_w, bbox_y]]
np_poly = np.array(poly).reshape((4,2))
polygons.append(Polygon(np_poly))
color.append(c)


ax = plt.gca()
ax.imshow(new_img)
p = PatchCollection(polygons, facecolor=color, linewidths=0, alpha=0.3)
ax.add_collection(p)
p = PatchCollection(polygons, facecolor='none', edgecolors=color, linewidths=2)
ax.add_collection(p)
```
